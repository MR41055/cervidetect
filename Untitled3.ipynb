When we train any machine learning model or any AI based model, we want to make sure that we take the entire dataset and divide them into three portions: Training, Testing, Cross-Validaion.

Training data is simply used to train the model. We feed the model with training data so that the model can learn relationships between the inputs and outputs. As the model is being trained, we want to make sure that the model is not overfitting the training data. Simply, as the model is being trained, we will use the cross-validation dataset and feed it into the model in order to make sure that the model is still generating good results on the validation dataset which the model has never seen before as part if the training data. As long as I am seeing error on both the training data and the validation data are going down, that's a good thing. That the model is either to generalize and not memorize. However, if we reach a state where we find that the error on the training data is going down and the error on the validation data is going up, that means the model is overfitting the training data and it starts to fail to generalize. Maybe now we need to intervene and essentially stop the training. After we finish training all the iterations, we take the train model and then we pass it along to the testing dataset and we feed the testing data. Then the model should be able to perform well.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Load datasets
cervical_cancer_data = pd.read_csv('/content/drive/MyDrive/cervical_cancer.csv')
fake_data = pd.read_csv('/content/drive/MyDrive/fake_data.csv')
```

```python
print(cervical_cancer_data.columns)
print(fake_data.columns)


# Assume 'target' is the target variable in both datasets
X_cc = cervical_cancer_data.drop('Biopsy', axis=1)
y_cc = cervical_cancer_data['Biopsy']

X_fake = fake_data.drop('Biopsy', axis=1)
y_fake = fake_data['Biopsy']
```

```python
# Split data into train and test sets
X_train_cc, X_test_cc, y_train_cc, y_test_cc = train_test_split(X_cc, y_cc, test_size=0.2, random_state=42)
X_train_fake, X_test_fake, y_train_fake, y_test_fake = train_test_split(X_fake, y_fake, test_size=0.2, random_state=42)

# Train and evaluate models
models = {
    'Logistic Regression': LogisticRegression()
}

results = []
```

```python
from sklearn.impute import SimpleImputer

# Replace '?' with NaN
X_train_cc.replace('?', np.nan, inplace=True)
X_test_cc.replace('?', np.nan, inplace=True)

# Convert columns to numeric
X_train_cc = X_train_cc.apply(pd.to_numeric)
X_test_cc = X_test_cc.apply(pd.to_numeric)

# Impute missing values with the mean
imputer = SimpleImputer(strategy='mean')
X_train_cc = imputer.fit_transform(X_train_cc)
X_test_cc = imputer.transform(X_test_cc)

# Train and evaluate models
for name, model in models.items():
    model.fit(X_train_cc, y_train_cc)
    y_pred_cc = model.predict(X_test_cc)
    acc_cc = accuracy_score(y_test_cc, y_pred_cc)
    print(f"Model: {name}, Accuracy: {acc_cc}")
```
